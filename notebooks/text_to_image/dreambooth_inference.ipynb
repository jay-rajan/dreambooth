```python
from diffusers import DiffusionPipeline
import torch

from diffusers import StableDiffusionPipeline
import torch


model_id = "/home/jrajan/dreambooth/models/pointling"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16).to("cuda")

prompt = "An image of a pntlng animation character on moon jumping"
image = pipe(prompt, num_samples=30, num_inference_steps=50, guidance_scale=7.5, width=512, height=512).images[0]
image
 


```

```python
import imageio
prompt = "An image of a pntlng animation swimming in the beach"
image = pipe(prompt, num_samples=30, num_inference_steps=50, guidance_scale=7.5, width=512, height=512).images[0]
imageio.imwrite("gen1.png", image)
image
```
